```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("/content/Titanicdataset.csv")
titanic.head()
```

```python
titanic.shape
```

```python
titanic.describe()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,4))

titanic.Survived.value_counts().plot(kind="bar",alpha=0.5)
```

```python
plt.scatter(titanic.Survived, titanic.Age, alpha=0.1)
plt.title("Age wrt Survived")
```

```python
for x in [1,2,3]:
    titanic.Age[titanic.Pclass == x].plot(kind="kde")
plt.title("Age wrt Pclass")
plt.legend(("1st","2nd","3rd"))
```
